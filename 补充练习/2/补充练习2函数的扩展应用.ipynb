### 1.模拟sorted函数，写出过程和原理

```python
list = [[2,45,34,2],[2,3,4],[23,4,2],[19,3,2],[45,1,2,124],[2,2,1]]

sorted(list)
# 过程如下
# 1) 首先按照 每个数组的第一个数字从小到大进行排序
# 2) 之后将 第一个元素相同的列表 按照第二个元素进行从小到大排序
# 3) 将上一次排序中被排列元素 相同的列表按照下一个元素进行从小到大排序
# 4) 重复上一步，直到有等值争议的数组都完成了排序，排序结束
```

## 2.比较使用range函数和使用enumerate函数的时间效率    
for i in range(len(seasons)) :#使用range函数控制索引   
   print(i, seasons[i])   
for i, e in enumerate(seasons) :#使用enumerate生成器对象   
   print(i, e)

```python
seasons = ["Spring", "Augest", "Autumn", "Winter"]
def range_test():
    for i in range(1, 10000):
        for i in range(len(seasons)):
            print(i, seasons[i])
%time range_test()
```

```python
def enumerate_test():
    for i in range(1, 10000):
        for i,e in enumerate(seasons):
            print(i, e)
%time enumerate_test()
```

### 3.   比较使用itemgetter函数和使用列表推导式的时间效率 
for i in range(100000):    
    itemgetter(0,1,3,6)(b) # ('zhang', 90)     

for i in range(100000):        
   [b[x] for x in range(len(b))  if x in [0,1,3,6]]


```python
import operator
b = ('zhang', 90, 'hello', 12, 'nihao', 124, 83)
def itemgetter_test():
    for i in range(100000):
        operator.itemgetter(0,1,3,6)(b)

%time itemgetter_test()

```

```python
def list_inference_test():
    for i in range(100000):
        [b[x] for x in range(len(b)) if x in [0,1,3,6]]
%time list_inference_test()
```

### 4.使用sorted函数排序，写出对a=[[20,1,3], [40,6,2], [10,5,0]]执行a_new=sorted(a)的排序过程

```python
a=[[20,1,3], [40,6,2], [10,5,0]]

a_new = sorted(a)

print(a_new)

#  排序过程如下
# 首先按照每一个子列表的第一个元素进行排序
# 第一轮结果为 [[10,5,0], [20,1,3], [40,6,2]]
# 不存在 第一个 元素相同的 列表，因此，排序结束
```

### 5.使用sorted函数排序，写出对a=[[20,1,3], [40,6,2], [10,5,0]]按第二个元素进行排序的代码

```python
a=[[20,1,3], [40,6,2], [10,5,0]]

a_new_sec = sorted(a, key=lambda x: x[1])
print(a_new_sec)
```

### 6.使用sorted函数排序，写出对a=[[20,1,3], [40,6,2], [10,5,0]]按三个元素的平均值大小进行排序的代码

```python
import statistics as st
a_new_avg = sorted(a, key=lambda x: st.mean([i for i in x]))
print(a_new_avg)
```

### 7.模拟函数装饰器，对现有自定义函数加入“X时X分X秒，启动XX函数”的提示

```python
import time
def MyFuncDecorator(func):
    def wrap_the_func(*args, **kwargs):
        t = time.localtime(time.time())
        print(f"[{t.tm_hour}时{t.tm_min}分{t.tm_sec}秒]启动{func.__name__}函数")
        result = func(*args, **kwargs)
        print("Your function has done!")
        return result
    return wrap_the_func

@MyFuncDecorator
def add(a, b):
    return a+b


print(add(1, 3))
```
