2.设计一个三维向量类，并实现向量的加法、减法以及向量与标量的乘法与浮点除法和向下取整除法运算。

```python
import numpy as np
class Vector3D():
    def add(self, a, b):
        if self.__check_vector(a, b):
            return np.array([a[i][j][m] + b[i][j][m] for i in range(len(a)) for j in range(len(a[0])) for m in range(len(a[0][0]))]).reshape(np.array(a).shape)
        else:
            self.wisper()
    def sub(self, a, b):
        if self.__check_vector(a, b):
            return np.array([a[i][j][m] - b[i][j][m] for i in range(len(a)) for j in range(len(a[0])) for m in range(len(a[0][0]))]).reshape(np.array(a).shape)
        else:
            self.wisper()
    def mul(self, a, b):
        if self.__check_vector(a, b):
            return np.array([a[i][j][m] * b[i][j][m] for i in range(len(a)) for j in range(len(a[0])) for m in range(len(a[0][0]))]).reshape(np.array(a).shape)
        else:
            self.wisper()
    def div(self, a, b):
        if self.__check_vector(a, b):
            return np.array([float(a[i][j][m]) / float(b[i][j][m]) for i in range(len(a)) for j in range(len(a[0])) for m in range(len(a[0][0]))]).reshape(np.array(a).shape)
        else:
            self.wisper()
    def div_round(self, a, b):
        if self.__check_vector(a, b):
            return np.array([float(a[i][j][m]) // float(b[i][j][m]) for i in range(len(a)) for j in range(len(a[0])) for m in range(len(a[0][0]))]).reshape(np.array(a).shape)
        else:
            self.wisper()
    def wisper(self):
        print("The length of the vector is incompatible")
    def __check_vector(self, a, b):
        a_np = np.array(a)
        b_np = np.array(b)
        if a_np.shape != b_np.shape:
            return False
        if len(a_np.shape) != 3 and len(b_np.shape) != 3:
            return False
        return True

vc = Vector3D()
vector_1 = np.arange(24).reshape(2,3,4)
vector_2 = np.ones([2,3,4])
print(f"The add:{vc.add(vector_1, vector_2)}\nThe sub:{vc.sub(vector_1, vector_2)}\nThe mul:{vc.mul(vector_1, vector_2)}\nThe div:{vc.div(vector_1, vector_2)}\nThe div_round:{vc.div_round(vector_1, vector_2)}")
```
